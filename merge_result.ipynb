```python
from pathlib import Path
import pandas as pd
```

```python
anti_inflam = Path("./result-anti-inflam.csv")
df_anti_inflam = pd.read_csv(anti_inflam)
df_anti_inflam
```

```python
go0006281 = Path("./pqqgnn/GO0006281/predictions.csv")
df_go0006281 = pd.read_csv(go0006281)
df_go0006281
```

```python
go0072593 = Path("./pqqgnn/GO0072593/predictions.csv")
df_go0072593 = pd.read_csv(go0072593)
df_go0072593
```

```python
go0098869 = Path("./pqqgnn/GO0098869/predictions.csv")
df_go0098869 = pd.read_csv(go0098869)
df_go0098869
```

```python
merged_df = df_anti_inflam.merge(df_go0006281, on='CID', how='outer')\
                          .merge(df_go0072593, on='CID', how='outer')\
                          .merge(df_go0098869, on='CID', how='outer')
merged_df
```

```python
merged_df.to_csv("merged_result.csv", index=False)
```
